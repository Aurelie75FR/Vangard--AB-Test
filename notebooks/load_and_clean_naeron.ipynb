```python
df = df_final_web_data.join(df_final_experiment_clients.set_index('client_id'), on='client_id')
df.head()
```

```python
df_control_web_data = df[df['Variation'] == 'Control']
df_control_web_data = df_control_web_data.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
df_control_web_data = df_control_web_data.reset_index(drop=True)
df_control_web_data.head()
```

```python
df_test_web_data = df[df['Variation'] == 'Test']
df_test_web_data = df_test_web_data.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
df_test_web_data = df_test_web_data.reset_index(drop=True)
df_test_web_data.head()
```

```python
df_control_web_data[df_control_web_data['process_step'] == 'start'] = df_control_web_data[df_control_web_data['process_step'] == 'start'].drop_duplicates(subset=['client_id', 'process_step'], keep='last')
df_control_web_data[df_control_web_data['process_step'] == 'confirm'] = df_control_web_data[df_control_web_data['process_step'] == 'confirm'].drop_duplicates(subset=['client_id', 'process_step'], keep='last')
df_control_web_data = df_control_web_data.dropna(how='any')
df_control_web_data = df_control_web_data.reset_index(drop=True)
df_control_web_data.head(20)
```

```python
df_test_web_data[df_test_web_data['process_step'] == 'start'] = df_test_web_data[df_test_web_data['process_step'] == 'start'].drop_duplicates(subset=['client_id', 'process_step'], keep='last')
df_test_web_data[df_test_web_data['process_step'] == 'confirm'] = df_test_web_data[df_test_web_data['process_step'] == 'confirm'].drop_duplicates(subset=['client_id', 'process_step'], keep='last')
df_test_web_data = df_test_web_data.dropna(how='any')
df_test_web_data = df_test_web_data.reset_index(drop=True)
df_test_web_data.head(20)
```
