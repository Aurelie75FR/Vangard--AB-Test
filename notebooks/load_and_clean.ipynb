```python
#Importing the libraries needed for this notebook 
import pandas as pd
import yaml
```

```python
#Importing raw data files from yaml file containing relative paths
with open('../config.yaml', 'r') as file:
    data_path = yaml.safe_load(file)

#Importing raw data file containing the user profile     
df_final_demo = pd.read_csv(data_path['raw_user_profile']['file'], sep=',')
#Importing raw data file containing the ID of the people in the experiment 
df_final_experiment_clients = pd.read_csv(data_path['raw_roster']['file'], sep=',')
#Importing raw data file containing the first part of digital footprints
df_final_web_data_pt_1 = pd.read_csv(data_path['raw_footprint_1']['file'], sep=',')
#Importing raw data file containing the second part of digital footprints
df_final_web_data_pt_2 = pd.read_csv(data_path['raw_footprint_2']['file'], sep=',')
```

```python
#Overview of DataFrame
df_final_web_data_pt_2.head()
```

```python
#Concatenating both parts of the DataFrame into a big one containing all the digital footprints
df_final_web_data = pd.concat([df_final_web_data_pt_1, df_final_web_data_pt_2])
df_final_web_data.head()
```

```python
#Converting the date_time column to a datetime object, creating two columns one for the date and the other for the hour
df_final_web_data['date_time'] = pd.to_datetime(df_final_web_data['date_time'], format="ISO8601")
df_final_web_data['date'] = df_final_web_data['date_time'].dt.date
df_final_web_data['time'] = df_final_web_data['date_time'].dt.time
```

```python
#Overview of DataFrame
df_final_web_data.head()
```

```python
#Renaming confirm step to validate. That way, it's easier to sort the DataFrame in the right order (alphabetical but also chronological)
df_final_web_data.loc[df_final_web_data['process_step'] == 'confirm', 'process_step'] = 'validate'
```

```python
#Sorting the DataFrame by client_id, visit_id, process_step and finally date_time
df_final_web_data.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
df_final_web_data.head()
```

```python
#Adding a column containing the group the client is in for the experiment
df = df_final_web_data.join(df_final_experiment_clients.set_index('client_id'), on='client_id')
#Converting to lowercase both column name and values
df['Variation'] = df['Variation'].str.lower()
df = df.rename(columns={'Variation' : 'variation'})
df.head()
```

```python
#Splitting dataset into one with only people from control group
df_control_web_data = df[df['variation'] == 'control']
df_control_web_data = df_control_web_data.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
df_control_web_data = df_control_web_data.reset_index(drop=True)
df_control_web_data.head()
```

```python
#And the other with people from the test group
df_test_web_data = df[df['variation'] == 'test']
df_test_web_data = df_test_web_data.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
df_test_web_data = df_test_web_data.reset_index(drop=True)
df_test_web_data.head()
```

```python
#We drop lines where there is a nan
df_control_web_data = df_control_web_data.dropna(how='any')
#Cast client_id to integer
df_control_web_data['client_id'] = df_control_web_data['client_id'].astype(int)
#Cast datetime to datetime object
df_control_web_data['date_time'] = pd.to_datetime(df_control_web_data['date_time'], format="ISO8601")
df_control_web_data = df_control_web_data.reset_index(drop=True)
df_control_web_data.head()
```

```python
#Same cleaning steps on the other dataset
df_test_web_data = df_test_web_data.dropna(how='any')
df_test_web_data['client_id'] = df_test_web_data['client_id'].astype(int)
df_test_web_data['date_time'] = pd.to_datetime(df_test_web_data['date_time'], format="ISO8601")
df_test_web_data = df_test_web_data.reset_index(drop=True)
df_test_web_data.head()
```

```python
#Exporting test group data to a csv file
df_test_web_data.to_csv('../data/clean/test_group_clean.csv', index=False)
```

```python
#Exporting control group data to a csv file
df_control_web_data.to_csv('../data/clean/control_group_clean.csv', index=False)
```

```python
#Concatenating and sorting again to have only the people who were part of the experiment
df_web_data_complete = pd.concat([df_test_web_data, df_control_web_data])
df_web_data_complete = df_web_data_complete.reset_index(drop=True)
df_web_data_complete = df_web_data_complete.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
```

```python
#Exporting test and control group data to the same csv file
df_web_data_complete.to_csv('../data/clean/complete_clean.csv', index=False)
```

```python
#Renaming columns of the user profile dataset
df_final_demo.columns = ['client_id', 'tenure_yr', 'tenure_mth', 'age',
       'gender', 'num_accounts', 'bal', 'calls_6_mth', 'logons_6_mth']
```

```python
#Only keeping the people that are part of the experiment
df_final_demo = df_final_demo[df_final_demo['client_id'].isin(df['client_id'])]
df_final_demo.head()
```

```python
#DataFrame overview
df_final_demo.head()
```

```python
#Dropping the tenure column in year because we already have tenure in months
df_final_demo = df_final_demo.drop(columns='tenure_yr')
```

```python
#Filling nan values with median tenure in months
df_final_demo['tenure_mth'] = df_final_demo['tenure_mth'].fillna(value=df_final_demo['tenure_mth'].median())
```

```python
#Filling nan values with U for unknown
df_final_demo['gender'] = df_final_demo['gender'].fillna(value='U')
```

```python
df_final_demo.describe()
```

```python
#Filling nan values with mean age
df_final_demo['age'] = df_final_demo['age'].fillna(value=int(df_final_demo['age'].mean()))
```

```python
#Filling nan values for these columns with median 
df_final_demo['num_accounts'] = df_final_demo['num_accounts'].fillna(value=df_final_demo['num_accounts'].median())
df_final_demo['calls_6_mth'] = df_final_demo['calls_6_mth'].fillna(value=df_final_demo['calls_6_mth'].median())
df_final_demo['logons_6_mth'] = df_final_demo['logons_6_mth'].fillna(value=df_final_demo['logons_6_mth'].median())
df_final_demo['bal'] = df_final_demo['bal'].fillna(value=df_final_demo['bal'].median())
```

```python
#Casting to the right types of data
df_final_demo['tenure_mth'] = df_final_demo['tenure_mth'].astype(int)
df_final_demo['age'] = df_final_demo['age'].astype(int)
df_final_demo['num_accounts'] = df_final_demo['num_accounts'].astype(int)
df_final_demo['calls_6_mth'] = df_final_demo['calls_6_mth'].astype(int)
df_final_demo['logons_6_mth'] = df_final_demo['logons_6_mth'].astype(int)
```

```python
#Exporting the dataset to a csv file
df_final_demo.to_csv('../data/clean/final_demo_clean.csv', index=False)
```

```python
#Combining everything into one big data set
big_df = df_web_data_complete.join(df_final_demo.set_index('client_id'), on='client_id')
big_df = big_df.reset_index(drop=True)
big_df = big_df.sort_values(by=['client_id', 'visit_id', 'process_step', 'date_time'])
```

```python
#Exporting the big dataset to a csv file
big_df.to_csv('../data/clean/all_clean.csv', index=False)
```
